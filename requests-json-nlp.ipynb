1. The following code accesses the [lyrics.ovh](https://lyricsovh.docs.apiary.io/#reference/0/lyrics-of-a-song/search) public api, searches for the lyrics of a song, and stores it in a dictionary object.  Write the resulting json to a file (either a JSON file or a pickle file; you choose). You will read in the contents of this file for future questions so we do not need to frequently access the API.

## Question #1 Response & Output:

```python
import requests
import json
```

```python

def get_lyrics(song, artist):
    url = f"https://api.lyrics.ovh/v1/{artist}/{song}"
    response = requests.get(url)
    if response.status_code == 200:
        return response.json()
    else:
        return {"error": "Lyrics not found"}

song = "Party in the USA"
artist = "Miley Cyrus"
lyrics_data = get_lyrics(song, artist)

if "lyrics" in lyrics_data:
    lyrics = lyrics_data["lyrics"]
    print("Lyrics of 'Party in the USA' by Miley Cyrus:\n")
    
    # Split the lyrics by newline characters
    lines = lyrics.split('\n')
    
    # Print each line with single spacing
    for line in lines:
        # Skip empty lines
        if line.strip():
            print(line.strip())

    # Writing the data to a JSON file
    with open("party_in_the_usa_lyrics.json", "w") as json_file:
        json.dump(lyrics_data, json_file)

    print("\nLyrics data written to party_in_the_usa_lyrics.json")
else:
    print("Lyrics not found.")

```

Print the polarity score of the sentiment analysis. ## Question #2 Response & Output:

```python
import spacy
from spacytextblob.spacytextblob import SpacyTextBlob

nlp = spacy.load("en_core_web_sm")
nlp.add_pipe("spacytextblob")

# Reading the contents of the JSON file
with open("party_in_the_usa_lyrics.json", "r") as json_file:
    lyrics_data = json.load(json_file)

# Extracting lyrics
lyrics = lyrics_data["lyrics"]

# Printing the lyrics
print("Lyrics of 'Party in the USA' by Miley Cyrus:\n")
print(lyrics)
print()

# Performing sentiment analysis
doc = nlp(lyrics)
polarity_score = doc._.polarity

# Printing the polarity score
print("Polarity Score:", polarity_score)
```

```python
# Given that the polarity score is 0.2906, which falls closer to 0 than to 1, the lyrics of "Party in the USA" by Miley Cyrus have a slightly positive connotation.
# Although the polarity score is positive, it's not extremely high, indicating that while there are elements of positivity in the lyrics, they may also contain neutral or slightly less positive sentiments. Therefore, while the overall sentiment is positive, it's not overwhelmingly so.
```

3. Write a function that takes an artist, song, and filename, accesses the lyrics.ovh api to get the song lyrics, and writes the results to the specified filename.  Test this function by getting the lyrics to any four songs of your choice and storing them in different files.

## Question #3 Response & Output:

```python
def get_lyrics_and_save(artist, song, filename):
    url = f"https://api.lyrics.ovh/v1/{artist}/{song}"
    response = requests.get(url)
    
    if response.status_code == 200:
        lyrics_data = response.json()
        with open(filename, "w") as json_file:
            json.dump(lyrics_data, json_file)
        print(f"Lyrics for '{song}' by {artist} saved to {filename}")
    else:
        print(f"Failed to retrieve lyrics for '{song}' by {artist}")

# Getting lyrics of 4 songs
get_lyrics_and_save("Ed Sheeran", "Shape of You", "shape_of_you_lyrics.json")
get_lyrics_and_save("Morgan Wallen", "Last Night", "last_night_lyrics.json")
get_lyrics_and_save("Luke Combs", "When It Rains It Pours", "when_it_rains_it_pours_lyrics.json")
get_lyrics_and_save("Taylor Swift", "Bad Blood", "bad_blood_lyrics.json")

```

4. Write a function that takes the name of a file that contains song lyrics, loads the file, performs sentiment analysis, and returns the polarity score.  Use this function to print the polarity scores (with the name of the song) of the three files you created in question 3.  Does the reported polarity match your understanding of the song's lyrics? Why or why not do you think that might be?  Answer the questions in either a comment in the code cell or a markdown cell under the code cell.

## Question #4 Response & Output:

```python
def analyze_sentiment(filename):
    nlp = spacy.load("en_core_web_sm")
    nlp.add_pipe("spacytextblob")
    
    # Reading the contents of the file
    with open(filename, "r") as json_file:
        lyrics_data = json.load(json_file)
    
    # Extracting lyrics
    lyrics = lyrics_data["lyrics"]
    
    # Performing sentiment analysis
    doc = nlp(lyrics)
    polarity_score = doc._.polarity
    return polarity_score

# Test the function by analyzing the sentiment of the files created in question 3
files = ["shape_of_you_lyrics.json", "last_night_lyrics.json", "when_it_rains_it_pours_lyrics.json", "bad_blood_lyrics.json"]
song_titles = ["Shape of You", "Last Night", "When It Rains It Pours", "Bad Blood"]

for file, title in zip(files, song_titles):
    # Analyze sentiment and print polarity score for each song
    polarity_score = analyze_sentiment(file)
    print(f"Polarity score for '{title}': {polarity_score}")

```

```python
# "Shape of You": 0.322 - This indicates a moderately positive sentiment. The song "Shape of You" by Ed Sheeran is generally upbeat and romantic, which aligns with the calculated positive sentiment score.

# "Last Night": 0.053 - This score is slightly positive but close to neutral. The song "Last Night" by Morgan Wallen may have a mix of emotions, potentially including some positive elements but also some more neutral or introspective themes, which could explain the close-to-neutral score.

# "When It Rains It Pours": -0.025 - This score is slightly negative but also close to neutral. The song "When It Rains It Pours" by Luke Combs has elements of overcoming challenges and turning bad situations into positive ones, but it may also contain some negative themes, such as setbacks or challenges.

# "Bad Blood": -0.268 - This score indicates a moderately negative sentiment. The song "Bad Blood" by Taylor Swift is known for its confrontational and negative themes, which often revolve around conflict and betrayal. Therefore, the negative sentiment score aligns with the tone of the song.
```
